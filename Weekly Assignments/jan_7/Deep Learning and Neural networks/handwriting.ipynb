```python
import cv2
import numpy as np
```

```python
digits=cv2.imread("/home/sesna/sesna/DL_proj/project/digits1.png",cv2.IMREAD_GRAYSCALE)
test_digits=cv2.imread("/home/sesna/sesna/DL_proj/project/test_digits1.png",cv2.IMREAD_GRAYSCALE)
print(test_digits)
print(len(test_digits))
print(len(digits))
```

```python
test_digits = np.vsplit(test_digits,50)  # splitting the image into 50 equal parts of 20 by 20 pixels
print(test_digits)
```

```python
test_cells=[]
for d in test_digits:
    d=d.flatten()     # converting to one dim array
    test_cells.append(d)
test_cells=np.array(test_cells,dtype=np.float32)


test_cells
```

```python
#cv2.imshow('image',test_cells[1])
```

```python
rows=np.vsplit(digits,50)
cells=[]

for row in rows:
    elem=np.hsplit(row,50)
    for i in elem:
        i=i.flatten()
        cells.append(i)

cells
```

```python
cells=np.array(cells,dtype=np.float32)   # converting to same dtype
cells
```

```python
k=np.arange(10)     # creating a numpy array contaning numbers 0 to 9
cells_labels=np.repeat(k,250)  # creating labels for each cells 
cells_labels
```

```python
knn=cv2.ml.KNearest_create()
knn.train(cells,cv2.ml.ROW_SAMPLE,cells_labels)   # (training data,telling the samples are in  row,lebals corresponding to each sample)


ret,result,neighboursv,dist=knn.findNearest(test_cells,k=20) 
```
